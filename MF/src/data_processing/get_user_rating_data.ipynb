이를 위해 전에 수집한 review데이터에 유저이름과 평점있는데 이를 이용하기로 결정

```python
import modules as m 

review_df = m.load_pickle('C:/Users/Hi/Desktop/recsys-movie/MF/data/scraping_results/results_1.pickle')
```

user title rating. 이렇게 3개를 쭉 수집해야한다.
이를 위해서는    
1. user, title, rating을 구하고 
2. 이를 pivot table로 만든다.

```python
# user, title, rating 구하기
review_list = []
for review in review_df:
    title_id = review[0]
    tmp_df = review[1]
    for i in range(len(tmp_df)):
      tmp_rating = tmp_df.iloc[i, 1]
      tmp_user = tmp_df.iloc[i, 3]
      review_list.append([tmp_user, title_id, tmp_rating])
```

```python
# pivot table 만들기
import pandas as pd

user_title_rating_df = pd.DataFrame(review_list, columns=['user', 'title', 'rating'])
user_title_rating_df
```

```python
pivot_table = user_title_rating_df.pivot_table(index='user', columns='title', values='rating')
```

```python
pivot_table
```

```python
pivot_table.shape
```

한 사람당 평균 몇 개의 영화에 평가를 했는지 확인해보자

```python
count_list = []

for i in range(len(pivot_table)):
  cnt_rating = len(pivot_table.iloc[i, :].dropna())
  count_list.append(cnt_rating)
```

```python
import numpy as np

np.mean(count_list)
```

영화 개수가 16000개 정도인데 한 사람당 평균 1.4개를 했다면 이를 그대로 사용할 수 없겠다.

```python
np.max(count_list)
```

```python
np.quantile(count_list, 0.99)
```

상위 1%가 9개, 최대 391개. 유저가 평가한 값이 너무 적다.

```python
len([count for count in count_list if count >=10])

# 평가개수가 10개가 넘는 유저의 개수는 386
```

```python
title_cnt_list = []

for title in pivot_table:
  title_cnt_list.append(len(pivot_table.loc[:, title].dropna()))
```

```python
np.mean(title_cnt_list), np.quantile(title_cnt_list, 0.8), np.max(title_cnt_list)
```

영화는 평균 6.5개의 rating을 받았고, 0.8 quantile에서는 11개, 최대 25개의 평가를 받았다.   
-> 당연하구나. 리뷰 받은 개수와 일치해야지

```python
np.mean(count_list) / pivot_table.shape[0] * 100, np.mean(title_cnt_list) / pivot_table.shape[1] * 100
```

평가되어있는 비율이 0.003%, 0.06%로 이대로는 사용할 수 없다고 판단된다.   
-> 그래도 전체의 10%정도는 평가가 되어있으면 좋겠다

동적 크롤링으로 리뷰 개수 늘렸고, 4000개정도로 영화 개수 줄여서 다시 데이터 긁어옴

```python
import modules as m 

review_df = m.load_pickle('C:/Users/Hi/Desktop/recsys-movie/MF/data/scraping_results/results_2.pickle')

# user, title, rating 구하기
review_list = []
for review in review_df:
    title_id = review[0]
    tmp_df = review[1]
    for i in range(len(tmp_df)):
      tmp_rating = tmp_df.iloc[i, 1]
      tmp_user = tmp_df.iloc[i, 3]
      review_list.append([tmp_user, title_id, tmp_rating])
```

```python
import pandas as pd 

user_title_rating_df = pd.DataFrame(review_list, columns=['user', 'title', 'rating'])
user_title_rating_df
```

```python
pivot_table = user_title_rating_df.pivot_table(index='user', columns='title', values='rating')
pivot_table
```

```python
import numpy as np

user_cnt = pivot_table.count(axis=1)
movie_cnt = pivot_table.count(axis=0)

user_rate = 0.99
movie_rate = 0.8

print(np.mean(user_cnt), np.quantile(user_cnt, user_rate), np.max(user_cnt))
print(np.mean(movie_cnt), np.quantile(movie_cnt, movie_rate), np.max(movie_cnt))
print(pivot_table.shape[0] * (1 - user_rate), pivot_table.shape[1] * (1 - movie_rate))
```

생각보다 더 평가한 횟수가 많지가 않네

```python
user_cnt = pivot_table.count(axis=1)
tmp_matrix = pivot_table.loc[(user_cnt > 10), :]

movie_cnt = tmp_matrix.count(axis=0)
tmp_matrix = tmp_matrix.loc[:, (movie_cnt > 3)]

print(f"{tmp_matrix.shape},\
  평가한 영화가 0개인 유저: {len(tmp_matrix.count(axis=1)[tmp_matrix.count(axis=1) < 1])},\
  평가한 유저가 0개인 영화 : {len(tmp_matrix.count(axis=0)[tmp_matrix.count(axis=0) < 1])}")
```

유저, 영화 2개 조건을 모두 만족시키게 필터링하는게 생각보다 되게 어렵네

```python
tmp_matrix.count(axis=1).mean(), tmp_matrix.count(axis=0).mean()  
```
