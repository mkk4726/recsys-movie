```python
# 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

cosine_matrix = cosine_similarity(rating_matrix.fillna(0))
cosine_df = pd.DataFrame(cosine_matrix, index=rating_matrix.index, columns=rating_matrix.index)
cosine_df
```

이웃의 가중평균으로 영화를 예측.  

```python
user_id = 'A_Different_Drummer'; movie_id = 'tt10062314'

rating_user = rating_matrix.loc[:, movie_id].dropna().index

neighbor = cosine_df.loc[user_id][rating_user].sort_values(ascending=False)[1:] # 가까운 애들에서 평가한 애들만 남겨야됨
neighbor
```
